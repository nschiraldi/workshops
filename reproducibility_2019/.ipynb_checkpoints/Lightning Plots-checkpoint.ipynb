```python
%matplotlib inline 

import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.ticker as mticker
import cartopy.crs as ccrs
import cartopy.mpl.ticker as cticker
import cartopy.feature as cfeature
```

Load in the data, process the datetimes for easier indexing of the data. It takes a while, but it's worth it.

```python
df=pd.read_csv(
    '9303.txt',
    header=None,
    names=['date','time','lat','lon','strike','num'],
    delim_whitespace=True,
    parse_dates={'timestamp': [0,1]},
    infer_datetime_format=True
).set_index('timestamp')
```

# Fig. 7

```python
fig = plt.figure(figsize=(8,8))

#-- 00 - 03 --#
ax = plt.subplot(2,2,1,projection=ccrs.PlateCarree())
ax.set_extent([-109.37, -87.5, 20.26, 37.37], ccrs.PlateCarree())
ax.add_feature(cfeature.STATES,linewidth=0.5,edgecolor='lightgrey')
ax.add_feature(cfeature.COASTLINE,linewidth=0.5)
ax.scatter(
    df.loc['1993-03-12 00:00:00':'1993-03-12 03:00:00',['lon']],
    df.loc['1993-03-12 00:00:00':'1993-03-12 03:00:00',['lat']],
    s=1,
    marker='o',
    transform=ccrs.PlateCarree(),
)
ax.set_title('a. Flashes: {0} | Valid: 12/00-12/03'.format(df.loc['1993-03-12 00:00:00':'1993-03-12 03:00:00','num'].sum()))

#--03 - 06 --#
ax = plt.subplot(2,2,2,projection=ccrs.PlateCarree())
ax.set_extent([-109.37, -87.5, 20.26, 37.37], ccrs.PlateCarree())
ax.add_feature(cfeature.STATES,linewidth=0.5,edgecolor='lightgrey')
ax.add_feature(cfeature.COASTLINE,linewidth=0.5)
ax.scatter(
    df.loc['1993-03-12 03:00:00':'1993-03-12 06:00:00',['lon']],
    df.loc['1993-03-12 03:00:00':'1993-03-12 06:00:00',['lat']],
    s=1,
    marker='o',
    transform=ccrs.PlateCarree(),
)
ax.set_title('b. Flashes: {0} | Valid: 12/03-12/06'.format(df.loc['1993-03-12 03:00:00':'1993-03-12 06:00:00','num'].sum()))

#-- 06 - 09 --#
ax = plt.subplot(2,2,3,projection=ccrs.PlateCarree())
ax.set_extent([-109.37, -87.5, 20.26, 37.37], ccrs.PlateCarree())
ax.add_feature(cfeature.STATES,linewidth=0.5,edgecolor='lightgrey')
ax.add_feature(cfeature.COASTLINE,linewidth=0.5)
ax.scatter(
    df.loc['1993-03-12 06:00:00':'1993-03-12 09:00:00',['lon']],
    df.loc['1993-03-12 06:00:00':'1993-03-12 09:00:00',['lat']],
    s=1,
    marker='o',
    transform=ccrs.PlateCarree(),
)
ax.set_title('b. Flashes: {0} | Valid: 12/06-12/09'.format(df.loc['1993-03-12 06:00:00':'1993-03-12 09:00:00','num'].sum()))

# -- 09 - 12 --#
ax = plt.subplot(2,2,4,projection=ccrs.PlateCarree())
ax.set_extent([-109.37, -87.5, 20.26, 37.37], ccrs.PlateCarree())
ax.add_feature(cfeature.STATES,linewidth=0.5,edgecolor='lightgrey')
ax.add_feature(cfeature.COASTLINE,linewidth=0.5)
ax.scatter(
    df.loc['1993-03-12 09:00:00':'1993-03-12 12:00:00',['lon']],
    df.loc['1993-03-12 09:00:00':'1993-03-12 12:00:00',['lat']],
    s=1,
    marker='o',
    transform=ccrs.PlateCarree(),
)
ax.set_title('b. Flashes: {0} | Valid: 12/09-12/12'.format(df.loc['1993-03-12 09:00:00':'1993-03-12 12:00:00','num'].sum()))
plt.subplots_adjust(hspace=0.01,wspace=0.1)
```
